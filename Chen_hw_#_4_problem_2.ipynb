$V_{\mathrm{eff}}\left(r\right)=\frac{L^{2}}{2\mu r^{2}}-V_{0}\exp\left(-\lambda^{2}r^{2}\right)$

Let $L'=\frac{\lambda L}{\sqrt{\mu V_{0}}}$ and $r'=\lambda r$,

$V_{\mathrm{eff}}\left(r'\right)=\frac{V_{0}L'^{2}}{2r'^{2}}-V_{0}\exp\left(-r'^{2}\right)$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from scipy import optimize
def Veff(r,l,V0):
    return V0*(l**2)/(2*r**2)-V0*np.exp(-(r)**(2))
    
```

```python
r=np.linspace(0.05,15,1000)
plt.plot(r,Veff(r,0.1,1))
#plt.plot(r,Veff(r,3,10))
plt.xlabel('$\lambda r$',fontsize=10)
plt.ylabel('$V_{eff}$',fontsize=10)
plt.title("The case with parameter $L'=0.1, V_{0}=1.0$")
plt.show()

r=np.linspace(0.2,7,1000)
plt.plot(r,Veff(r,0.3,1))
#plt.plot(r,Veff(r,3,10))
plt.xlabel('$\lambda r$',fontsize=10)
plt.ylabel('$V_{eff}$',fontsize=10)
plt.title("The case with parameter $L'=0.3, V_{0}=1.0$")
plt.show()

r=np.linspace(0.3,6,1000)
plt.plot(r,Veff(r,0.5,1))
#plt.plot(r,Veff(r,3,10))
plt.xlabel('$\lambda r$',fontsize=10)
plt.ylabel('$V_{eff}$',fontsize=10)
plt.title("The case with parameter $L'=0.5, V_{0}=1.0$")
plt.show()

r=np.linspace(0.5,4,1000)
plt.plot(r,Veff(r,0.7,1))
plt.xlabel('$\lambda r$',fontsize=10)
plt.ylabel('$V_{eff}$',fontsize=10)
plt.title("The case with parameter $L'=0.7, V_{0}=1.0$")
plt.show()

r=np.linspace(0.8,5,1000)
plt.plot(r,Veff(r,1,1))
plt.xlabel('$\lambda r$',fontsize=10)
plt.ylabel('$V_{eff}$',fontsize=10)
plt.title("The case with parameter $L'=1, V_{0}=1.0$")
plt.show()

r=np.linspace(0.8,5,1000)
plt.plot(r,Veff(r,1.0405201900457776,1))
plt.xlabel('$\lambda r$',fontsize=10)
plt.ylabel('$V_{eff}$',fontsize=10)
plt.title("The case with parameter $L'=1.0405201900457776, V_{0}=1.0$")
plt.show()

r=np.linspace(0.8,5,1000)
plt.plot(r,Veff(r,1.05,1))
plt.xlabel('$\lambda r$',fontsize=10)
plt.ylabel('$V_{eff}$',fontsize=10)
plt.title("The case with parameter $L'=1.05, V_{0}=1.0$")
plt.show()
```

To find out the local minimum of effective potential, I dervative the dffective potential by $r'$

$\frac{\partial V_{\mathrm{eff}}\left(r'\right)}{\partial r'}=-\frac{V_{0}L'^{2}}{r'^{3}}+2r'V_{0}\exp\left(-r'^{2}\right)=0$

After rearranging, I got the

$0=-\frac{V_{0}L'^{2}}{2}+r'^{4}V_{0}\exp\left(-r'^{2}\right)$

Therefore, I can plot it to find out how many solution it has and also find out which one corresponds to the local minimun.

```python
def g(r,l): #To find out the local minimum of effective potential
    return (r**4)*np.exp(-r**2)-(l**2)/2
```

```python
r=np.linspace(0,10,1000)
plt.plot(r,g(r,0))

plt.plot(r,g(r,0.3))
plt.plot(r,g(r,0.5))
plt.plot(r,g(r,0.7))
plt.plot(r,g(r,0.9))
#plt.plot(r,g(r,1.2))
#plt.plot(r,g(r,1.5))
#plt.plot(r,Veff(r,3,10))
plt.show()
```

Based on the plot above and effective potential I plot, the solution which r is small one is correponding the local minimun. Therefore, I can use it to find out the radius of stable circular orbit. 

```python
optimize.root(g, [0.7,],args=(0.3), method='lm')

```
