The potential is

$V=-\frac{k}{r}+\frac{h}{r^{2}}$

The Lagrangian is 

$L=T-V=\frac{1}{2}m\left(\dot{r}^{2}+r^{2}\dot{\theta}^{2}\right)-V\left(r\right)$

The Eular Lagrange equation for $r$ is 

$m\ddot{r}-mr\dot{\theta}=-\frac{\partial V}{\partial r}=-\frac{k}{r^{2}}+\frac{2h}{r^{3}}$

The Eular Lagrange equation for $\theta$ is

$\frac{d}{dt}mr^{2}\dot{\theta}=0$

$mr^{2}\dot{\theta}=l$

After rearranging the Eular Lagrange equation I got, the equation becomes

$m\ddot{r}=-\frac{k}{r^{2}}+\frac{2mh+l^{2}}{mr^{3}}$

From the angular moment equation, I can use $\frac{d}{dt}=\frac{l}{mr^{2}}\frac{d}{d\theta}$ to write the equation above

$\frac{l}{r^{2}}\frac{d}{d\theta}\left(\frac{l}{mr^{2}}\frac{dr}{d\theta}\right)-\frac{l^{2}}{mr^{3}}=-\frac{k}{r^{2}}+\frac{2h}{r^{3}}$

Let $u=\frac{1}{r}$ and $du=-\frac{1}{r^{2}}dr$. The equation becomes

$\frac{d^{2}u}{d\theta^{2}}+\left(\frac{l^{2}+2hm}{l^{2}}\right)u-\frac{km}{l^{2}}=0$

One of the solution is

$u=\frac{km}{l^{2}+2hm}+A\cos\left[\sqrt{\frac{l^{2}+2hm}{l^{2}}}\left(\theta-\theta'\right)\right]$

The angular speed of orbital is

$\frac{2\pi\left(1-\sqrt{\frac{l^{2}}{l^{2}+2hm}}\right)}{\tau}=\dot{\Omega}$

By using the approximation $l^{2}\gg2hm$ , I can get

$\frac{2\pi hm}{l^{2}\tau}=\dot{\Omega}$




```python
import numpy as np
import matplotlib.pylab as plt
def r(theta,l,h,e,k,m):
    x=(l**2+2*h*m)/(l**2)
    y=np.sqrt(x)
    return 1/((k*m)/(l**2+2*h*m)+e*np.cos(y*(theta)))

theta = np.linspace(0, 2*np.pi, 10000)

ax = plt.subplot(111, projection='polar')
ax.plot(theta, r(theta,1.5,0.05,0.2,1,1))
ax.set_rmax(5)
ax.set_rticks([1.0, 2.0, 3.0, 4.0])  # less radial ticks
ax.grid(True)
plt.show()
```
